#### Calculo do coeficiente Angular da reta


```python
y = Moscow
x = datas
n = np.size(Moscow) # Par de dados
```

```python
x**2
```

```python
a = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
a
```

#### Cálculo do Coeficiente linear da reta


```python
b = np.mean(y) - a*np.mean(x)
b
```

```python
Y = a*x + b
np.linalg.norm(Moscow-Y)
```

```python
plt.plot(datas,Moscow)
plt.plot(x,Y)
plt.plot(41.5,41.5*a + b,'*r')
plt.plot(100,100*a + b,'*r') # Projeção no mês 100
```

```python
import pandas as pd

dados = pd.read_csv('alunos.csv',sep=",")
```

```python
dados.isnull().sum() #Verificando a quantidade de dados nulos
```

```python
dados = dados.fillna(0) #Substituindo os dados nulos por 0
dados
```

```python
nomes_a_remover = dados.query('Nome == "Alice" | Nome == "Carlos"').index # Buscando o index de alice e Carlos
nomes_a_remover
```

```python
dados.drop(nomes_a_remover,axis=0,inplace=True) #Removendo Alice e Carlos
```

```python
aprovados = dados["Aprovado"] == True
dados = dados[aprovados]
dados
```

```python
dados['Notas'] = dados['Notas'].replace(7.0,8.0) #Substituindo todas as notas 7 do dataframe pela nota 8
dados
```
